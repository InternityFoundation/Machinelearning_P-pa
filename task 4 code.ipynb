```python
names = ["sepal length" , "sepal width" , "petal length" , "petal width" , "class" ]
df = pd.read_csv("iris.data" , names = names )
```

```python
df.head()
```

```python
df.info()
'''The dataset contains the 'class' values in text format
and hence we need to change it to numerical values'''
```

```python
df["class"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['class'] = le.fit_transform(df['class'])
```

```python
'''Labels encoded in aphabetical order
   Iris-virginica  :  2
   Iris-versicolor :  1
   Iris-setosa     :  0 '''
df
```

```python
# Now dividing the dataset into X and Y(inputs and outputs)

X = df.drop(["class"] , axis = 1)
y = df["class"]
```

```python
#train and test split of dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4,test_size=0.20)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
